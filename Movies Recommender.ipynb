### Simple recommender

##### The value of m determines the movies that qualify to be in the chart and also, by being part of the formula, determines the final value of the score.
For our recommender, we will use the number of votes garnered by the 80th percentile movie as our value for m. In other words, for a movie to be considered in the rankings, it must have garnered more votes than at least 80% of the movies present in our dataset. Additionally, the number of votes garnered by the 80th percentile movie is used in the weighted formula described previously to come up with the value for the scores.

```python
m=data['vote_count'].quantile(0.95)
m
```

#### Another prerequisite that we want in place is the runtime. The length of the movie should not be equal to 0.

```python
datan=data[data['runtime']!=0]
```

```python
datan=data[data['vote_count']>=m]
```

```python
datan.shape
```

```python
C=data['vote_average'].mean()
C
```

```python
def movie_rating(r,m=m,C=C):
    v=r['vote_count']
    R=r['vote_average']
    return (v/v+m*R)+(m/v+m*C)
```

```python
# Compute the score using the weighted_rating function defined above
datan['score']=datan.apply(movie_rating,axis=1)
```

####consider three movies: A, B, and C. How can we mathematically prove that the plot of A is more similar to the plot of B than to that of C (or vice versa)? 

The first step toward answering these questions is to represent the bodies of text (henceforth referred to as documents) as mathematical quantities. This is done by representing these documents as vectors. In other words, every document is depicted as a series of n numbers, where each number represents a dimension and n is the size of the vocabulary of all the documents put together.

### Content based recommender

```python
df['tagline']=df['tagline'].fillna('')
df['overview']=df['overview'].fillna('')
df['description']=df['tagline']+df['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#removing the stopwords
tfidf=TfidfVectorizer(stop_words='english')

#constructing tfidf matrix
tfidf_matrix=tfidf.fit_transform(df['description'])
tfidf_matrix.shape
```

#### we have got a 22222 dimensional matrix

```python
#cosine similarity
from sklearn.metrics.pairwise import linear_kernel

cosine_sim=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
sim_score=list(enumerate(cosine_sim[idx]))
```

```python
sim_score=sorted(sim_score,key=lambda x:x[1],reverse=True)
top=sim_score[1:11]
```

```python
#create a function that takes in the movie name as argument
def content_based(title, cosine_sim=cosine_sim, df=df, indices=indices):
    #obtain the index of the movie
    idx=indices[title]
    #get the pairwise score and convert into list
    sim_score=list(enumerate(cosine_sim[idx]))
    #sort the movies based on the score
    sim_score=sorted(sim_score, key=lambda x:x[1],reverse=True)
    #get top 10 elements of the list
    top=sim_score[1:11]
    #get movie indices of the top 10 movies
    movie_indices=[i[0] for i in top]
    #get names of movies
    return df['title'].iloc[movie_indices]
    
    
```

```python
#Get recommendations for The Lion King
content_based('The Dark Knight Rises')
```

#### The algorithm suggest all the movies in the Batman Franchise. This shows that our recommender is working absolutely fine, because obviously a person who loves 'The dark knight rises' is a Batman or Christopher Nolan fan and the recommender shows other Batman and Nolan movies in the suggestions.

### Metadata based recommender 

#### We'll take into account other parameters such as director, genre, cast and keywords to build this moddel.

```python
df2=pd.read_csv('tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
df2.head()
```

#### we'll now join both the dataframes. 

```python
df2.rename(columns={'movie_id':'id'},inplace=True)
```

```python
def convert_int(x):
    try:
        return int(x)
    except:
        return np.nan
```

```python
df['id']=df['id'].apply(convert_int)

df=df[df['id'].notnull()]
```

```python
df.info()
```

```python
df=df.merge(df2,on='id')
df.head()
```

```python
features=['cast','crew','genres']
for feature in features:
    df[feature]=df[feature].apply(literal_eval)
```

```python
df.rename(columns={'title_x':'title'},inplace=True)
```

```python
def get_director(x):
    for crew_member in x:
        if crew_member['job']=='Director':
            return crew_member['name']
    return np.nan
```

```python
df['director']=df['crew'].apply(get_director)
```

```python
df['director']
```

```python
def get_cast(x):
    if isinstance(x,list):
        names= [i['name'] for i in x]
        if len(names)>3:
            names=names[:3]
        return names
    return []
```

```python
df['actor']=df['cast'].apply(get_cast)
```

```python
df['actor']
```

```python
df['genres']=df['genres'].apply(lambda x: x[:3])
```

```python
df['genres'].head()
```

```python
def data_clean(x):
    if isinstance(x,list):
        return [str.lower(i.replace(" ","")) for i in x]
    elif isinstance(x,string):
        return str.lower(x.replace(" ",""))
    else:
        return ''
```

```python
df['actors']=df['actor'].apply(data_clean)
```

```python
df['actors'].head()
```

```python
def create_soup(x):
    try:
        return ' '.join(x['actors'])+' '+' '.join(x['genres'])+' '+x['director']
    except:
        return  ' '.join(x['actors'])+' '+' '.join(x['genres'])
```

```python
df['soup']=df.apply(create_soup,axis=1)
```

#### Instead of using TF-IDFVectorizer, we will be using CountVectorizer.This is because using TF-IDFVectorizer will accord less weight to actors and directors who have acted and directed in a relatively larger number of movies.

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer(stop_words='english')

count_matrix=count.fit_transform(df['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2=cosine_similarity(count_matrix,count_matrix)
    
```

#### With the new reverse mapping constructed and the similarity scores computed, we can reuse the content_based function defined in the previous section by passing in cosine_sim2 as an argument. Let's now try out our new model by asking recommendations for the same movie, The Dark Knight Rises:

```python
content_based('The Dark Knight Rises',cosine_sim2,df,indices)
```

#### The recommendations given in this case are vastly different to the ones that our plot description-based recommender gave. We see that it has been able to capture more information than just batman. It has recommended us the prestige which is another masterpiece by the same director, nolan.
